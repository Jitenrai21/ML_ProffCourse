```python
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import matplotlib.pyplot as plt
```

```python
loaded_model = load_model('Number_Classification_NNModel.h5')
```

```python
def predict(model, path):
    image = load_img(path, color_mode='grayscale', target_size=(28,28))
    image = img_to_array(image).reshape(1,28,28,1).astype('float32')/255.0
    prediction = model.predict(image)
    return prediction.argmax()
```

```python
path = '7.png'
predicted = predict(loaded_model, path)
image = load_img(path, color_mode='grayscale', target_size=(28,28))
plt.imshow(image, cmap='gray')
plt.title(f"Predicted Digit: {predicted}")
plt.show()
```
